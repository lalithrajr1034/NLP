<h1>5. Chat word tratement</h1>

```python
# for this make a dictanary then apply that function to a data
```

```python
text = "i luv dis movie bt it ws too lng omg!!!"
```

```python
# we git this in git 
chat_words = {
    "luv"  : "love",
    "dis"  : "this",
    "bt"   : "but",
    "ws"   : "was",
    "lng"  : "long",
    "omg"  : "oh my god",
    "lol"  : "laughing out loud",
    "brb"  : "be right back",
    "idk"  : "i don't know",
    "imo"  : "in my opinion",
    "tbh"  : "to be honest",
    "ngl"  : "not gonna lie",
    "u"    : "you",
    "r"    : "are",
    "ur"   : "your",
    "cant" : "cannot",
    "wont" : "will not",
    "gr8"  : "great",
    "b4"   : "before",
    "2"    : "to",
    "4"    : "for",
}
```

```python
def treat_chat_words(text):
    words = text.lower().split()                       
    cleaned = [chat_words.get(w, w) for w in words]    
    return ' '.join(cleaned)                            

text = "omg i luv dis movie bt it ws 2 lng"
treat_chat_words(text)
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()

def correct_spellings(text):
    words = text.split()
    corrected = [spell.correction(w) for w in words]
    return ' '.join(corrected)

text = "I luv thiss moviee it was amazng"
correct_spellings(text)
```

<h1> 7. Remove Stop words</h1>

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word not in stop_words:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```
