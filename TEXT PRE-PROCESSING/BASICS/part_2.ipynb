<H1> Handline emojis</H1>

```python
# to remvoe emoji
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map
        "\U0001F700-\U0001F77F"  # alchemical
        "\U0001F780-\U0001F7FF"  # geometric
        "\U0001F800-\U0001F8FF"  # supplemental arrows
        "\U0001F900-\U0001F9FF"  # supplemental symbols
        "\U0001FA00-\U0001FA6F"  # chess etc.
        "\U0001FA70-\U0001FAFF"  # extended symbols
        "\U00002702-\U000027B0"  # dingbats
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)


text = "I love NLP 😍🔥🚀"
remove_emojis(text)
```

<h1>Stemming </h1>

```python
# Stemming is the process of reducing words to their root/base form by removing suffixes.
# running -> run
# played  -> play
```

```python
# To perfrom the stemming there are stemmers(algorithm) that is made with lingustic experts(or who had a knowledge of langueage)
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem_word(text):
    a = text.split()
    for i in range(len(a)):
        a[i] = ps.stem(a[i])
    return ' '.join(a)
```

```python
text = 'hi my names running is lalith'
stem_word(text)
```

```python
# 1️⃣ Stemming cuts words to root form but may produce incorrect words (e.g., studies → studi).
# 2️⃣ It does not understand meaning or grammar.
# 3️⃣ It can cause over-stemming and under-stemming problems.
# 4️⃣ Lemmatization returns correct base words using vocabulary and context (e.g., studies → study).
# 5️⃣ Therefore, lemmatization is preferred when accuracy is important, while stemming is used for speed.
```

```python
| Word    | Stemming 🌱  | Lemmatization 🧠  |
| ------- | -----------  | ----------------  |
| studies | studi ❌     | study ✅          |
| running | run ✅       | run ✅            |
| better  | better ❌    | good ✅           |
| went    | went ❌      | go ✅             |
```

<H1>Lemmatization</H1>

```python
# This will convert in to a meaning full word but it is slow but stemming is fast
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["studies", "running", "better", "went"]
for word in words:
    print(word, "→", lemmatizer.lemmatize(word,pos = 'v')) # here you can alter a part of spech
```
